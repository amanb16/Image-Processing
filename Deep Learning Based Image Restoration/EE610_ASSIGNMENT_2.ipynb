```python
    
""" ADDING NOISE TO THE IMAGES AND STORING LOCALLY"""   
from PIL import Image, ImageTk, ImageFilter, ImageEnhance, ImageOps
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.util import random_noise
outPath ="C:\\Users\\MK857\\OneDrive\\Documents\\MTP\\ee610\\train_noisyblurry"
inPath ="C:\\Users\\MK857\\OneDrive\\Documents\\MTP\\ee610\\train_blurred"

for imagePath in os.listdir(inPath):
    inputPath = os.path.join(inPath, imagePath)
    img = Image.open(inputPath)
    img=np.asarray(img)
    noise_img = random_noise(img, mode='s&p',amount=0.05)
    noise_img = np.array(255*noise_img, dtype = 'uint8')
    data = Image.fromarray(noise_img)
    image=data 
    fullOutPath = os.path.join(outPath, 'noisyblurry_'+imagePath)
    image.save(fullOutPath)
    #print(fullOutPath)
for imagePath in os.listdir(inPath):
    inputPath = os.path.join(inPath, imagePath)
    img = Image.open(inputPath)
    img=np.asarray(img)
    noise_img = random_noise(img, mode='gaussian')
    noise_img = np.array(255*noise_img, dtype = 'uint8')
    data = Image.fromarray(noise_img)
    image=data 
    fullOutPath = os.path.join(outPath, 'noisyblurry_'+imagePath)
    image.save(fullOutPath)
    print(fullOutPath)

    
```

```python
1)LOAD THE ORIGINAL IMAGES FROM LOCAL STORAGE 
2)EXTRACT THE PATCHES OF WINDOW SIZE W=13
3)TENSOR FLOW MODULE IS USED TO EXTRACT THE PATCHES 
4)A TOTAL OF 888 PATCHES OBTAINED FOR EACH IMAGE
5)WE TAKEN ONLY 200 PATCHES FOR EVERY IMAGE SO A TOTAL OF 40,000 PATCHES CAME FOR 200 IMAGES 
6)WE HAVE TAKEN CENTRAL PIXEL FROM EVERY PATCH AND STORED THEM IN AN ARRAY CALLED c_v
7)c_v ARRAY OBTAINED HERE IS ASSIGNED TO y SO WE CAN USE THIS AS A OUTPUT TO TRAIN ML MODEL
```

```python
"""TAKING CENTRAL VALUES FROM THE ORIGINAL PATCHES AND STORING THEM IN ARRAY c_v """
import pandas as pd
import matplotlib.pyplot as plt 
from PIL import Image, ImageTk, ImageFilter, ImageEnhance, ImageOps
import os
import cv2
import numpy as np
import tensorflow as tf

from sklearn.datasets import load_sample_image
from sklearn.feature_extraction import image
inPath ="C:\\Users\\MK857\\OneDrive\\Documents\\MTP\\ee610\\train"
k=[]
p=[]
cv=[]
for imagePath in os.listdir(inPath):
    inputPath = os.path.join(inPath, imagePath)
    img = Image.open(inputPath)
    img=np.asarray(img)
    image = tf.expand_dims(img,0)
    #plt.imshow(img)
    patches=tf.image.extract_patches(images=image,sizes=[1, 13, 13, 1],strides=[1, 13, 13, 1],rates=[1, 1, 1, 1],padding='VALID')
    patches=patches.numpy()
    patches=(patches.reshape(-1,13,13,3))
#patches =image.extract_patches_2d(img, (13, 13), max_patches=13,
          #random_state
    k.append(patches)
k=np.array(k)

s=[]
for i in range(200):
    t=k[i]
    S=[]
    for j in range(200):
        p=t[j][:][:][:]
        S.append(p)
    s.append(S)
s=np.array(s)
k=s
print(k.shape)
c_v=[]
for i in range(200):
    t=k[i]
    CV=[]
    for j in range(200):
        p=t[j][:][:][:]
        cv=p[6][6]
        CV.append(cv)
    c_v.append(CV)
c_v=np.array(c_v)
c_v=c_v.reshape(-1,3)
y=c_v
print("output is ready")
print("SHAPE OF OUTPUT CENTRAL PIXEL ARRAY",c_v.shape) 
```

```python
1)LOAD THE DEGRADED IMAGES FROM LOCAL STORAGE 
2)EXTRACT THE PATCHES OF WINDOW SIZE W=13
3)TENSOR FLOW MODULE IS USED TO EXTRACT THE PATCHES 
4)A TOTAL OF 888 PATCHES OBTAINED FOR EACH IMAGE
5)WE TAKEN ONLY 200 PATCHES FOR EVERY IMAGE SO A TOTAL OF 40,000 PATCHES CAME FOR 200 IMAGES 
6)WE STORED THESE ARRAY OF PATCHES IN x SO WE CAN USE THIS AS A input TO TRAIN ML MODEL
```

```python
""" TAKING DEGRADED IMAGES PATCHES AND STORING THEM IN ARRAY K"""
from PIL import Image, ImageTk, ImageFilter, ImageEnhance, ImageOps
import os
import cv2
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.datasets import load_sample_image
from sklearn.feature_extraction import image
inPath = "C:\\Users\\MK857\\OneDrive\\Documents\\MTP\\ee610\\train_noisyblurry"
k=[]
for imagePath in os.listdir(inPath):
    inputPath = os.path.join(inPath, imagePath)
    img = Image.open(inputPath)
    img=np.asarray(img)
    image = tf.expand_dims(img,0)
    #plt.imshow(img)
    patches=tf.image.extract_patches(images=image,sizes=[1, 13, 13, 1],strides=[1, 13, 13, 1],rates=[1, 1, 1, 1],padding='VALID')
    patches=patches.numpy()
    patches=(patches.reshape(-1,13,13,3))
#patches =image.extract_patches_2d(img, (13, 13), max_patches=13,
          #random_state
    k.append(patches)
k=np.array(k)
s=[]
for i in range(200):
    t=k[i]
    S=[]
    for j in range(200):
        p=t[j][:][:][:]
        S.append(p)
    s.append(S)
s=np.array(s)
k=s
#print(k.shape)
k=k.reshape(-1,507)
x=k
print("Shape of the degraded patches array :",x.shape)



```

```python
1)Splitting the x && y arrays in to training and testing sets 
2)Used sklearn module to do the spliting of data 
3)80% is used for training and 20% is used for testing 
```

```python
""" DATA SPLITTING FOR TRAINING && TESTING """
import seaborn as sns
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1) 
#print(x_train)
print("TRAINING && TESTING DATA SPLIT IS  DONE")
 
```

```python
""""                               DECISION TREE REGRESSION MODEL         """
```

```python
""""1)I SELECTED DECISION TREE REGRESSION MODEL HERE WITH THE FOLLOWING HYPER PARAMETERS 
criterion='mse'
splitter='best'
max_depth=None
min_samples_split=2
min_samples_leaf=1
min_weight_fraction_leaf=0.0
max_features=None
random_state=None
max_leaf_nodes=None
min_impurity_decrease=0.0
ccp_alpha=0.0 
2)MOST OF THESE ABOVE VALUES ARE PREDEFINED 
3)Trained the model using x_train,y_train 
4)Here we are training model to predict the clean central pixels when degraded image patches are given as input 
"""
```

```python
""" TRAINING DECISION TREE ML MODEL   """
from sklearn.tree import DecisionTreeRegressor
regressor =DecisionTreeRegressor(criterion='mse', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, ccp_alpha=0.0)
dt=regressor.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print("DecisionTree MODEL IS DONE")
```

```python
THE ERROR METRICS OF THE ABOVE DECISION TREE MODEL 
1)Used sklearn module 
```

```python
print("THESE ARE THE METRICS OF DECISION TREE MODEL ")
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
